# MIMIC-IV Radiology Note Analysis

```python
import pandas as pd
import os
import numpy as np

csv_path = os.path.expanduser("~/data/mimic/mimiciv-note/2.2/radiology.csv")
df = pd.read_csv(csv_path)
df
```

```python
def print_random_notes(df, n=5):
    random_indices = np.random.choice(df.index, n, replace=False)
    for idx in random_indices:
        print(f"--- Note ID: {df.loc[idx, 'note_id']} | Note Type: {df.loc[idx, 'note_type']} ---")
        print(f"{df.loc[idx, 'text']}\n\n")

print_random_notes(df)
```
